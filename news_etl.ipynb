```python
import pandas as pd
import requests
import sqlalchemy
```

```python
# Extract
API_KEY = "92431105e7494e548d75225c9b0deaa8"
COUNTRY = "ca"
r = requests.get(f"https://newsapi.org/v2/top-headlines?country={COUNTRY}&apikey={API_KEY}")
data = r.json()
print(f"status: {data['status']}, totalResults: {data['totalResults']}")
```

```python
# Data frame
authors = []
titles = []
descriptions = []
urls = []
urls_to_image = []
published_at = []
ids = []

if data["status"] == "ok":
    for article in data["articles"]:
        authors.append(article["author"])
        titles.append(article["title"])
        descriptions.append(article["description"])
        urls.append(article["url"])
        urls_to_image.append(article["urlToImage"])
        published_at.append(article["publishedAt"])
        #ids.append(str(uuid.uuid1()).split("-")[0])
else:
    print(f"Status: {r.status_code}")
```

```python
# news dictionary
news_dict = {
    'author': authors,
    'title': titles,
    'description': descriptions,
    'url': urls,
    'url_to_image': urls_to_image,
    'published_at': published_at
}
```

```python
# Load data
import sqlite3 as db

DATABASE_LOCATION = 'news_database.db'
DATABASE_ENGINE = f'sqlite:///{DATABASE_LOCATION}'

sqlite_connection = db.connect(DATABASE_LOCATION)
cursor = sqlite_connection.cursor()
engine = sqlalchemy.create_engine(DATABASE_ENGINE)
```

```python
sql_query = """
    CREATE TABLE IF NOT EXISTS news(
    author VARCHAR(200),
    title VARCHAR(1000),
    description VARCHAR(1000),
    url VARCHAR(500),
    url_to_image VARCHAR(500),
    published_at VARCHAR(100))
"""
```
